```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame creation for demonstration
data = {'business/farm in hhld': [1, 2, 1, -3, -2, -1, 1, 2, -1]}
df = pd.DataFrame(data)

# Descriptions for the pie chart values
descriptions = {
    "1": "Yes",
    "2": "No",
    "-3": "Refused",
    "-2": "Don't Know",
    "-1": "Blank"
}

# Plotting the pie chart
plt.figure(figsize=(8, 8))
pie_data = df['business/farm in hhld'].value_counts()
labels = [descriptions[str(index)] if str(index) in descriptions else index for index in pie_data.index]
wedges, texts, autotexts = plt.pie(pie_data, labels=labels, autopct='%1.1f%%', startangle=90)

# Setting the title and removing the y-label
plt.title('Business/Farm in Household Distribution')
plt.ylabel('')  # Hide y-label as it's unnecessary for pie charts

# Adding a legend
plt.legend(wedges, labels, title="Status", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```
